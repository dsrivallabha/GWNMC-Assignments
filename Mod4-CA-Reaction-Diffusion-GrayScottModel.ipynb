# Reaction Diffusion equation - Gray Scott model

```python
import numpy
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
n = 192

Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065 # Bacteria 1 

dh = 5./(n-1)

T = 1

dt = .9 * dh**2 / (4*max(Du,Dv))

nt = int(T/dt)
```

```python
def ftcs(U, V, n, Du, Dv,  F, k, nt, dt, dh):
    """Solves the reaction - diffusion equation with forward-time, centered scheme
    
    Parameters:
    ----------
    U,V: array of float
        Species concentration profiles
    n: int
        no of points in each direction
    F,k: int
        feeding and removal terms
    nt: int
        Number of time steps
    dt: float
        Time step size
    dh: float
        Mesh size
        
    Returns:
    -------
    U,V: array of float
        returns updated values of U,V
    
    """
    #print(n,Du, Dv, F,k,nt,dt,dh)
    bydhsqr = 1./(dh**2)
    for tstep in range(nt):  
        Un = U.copy() 
        Vn = V.copy() 
        
        #print(tstep)
        for j in range(1,n-1):
            for i in range(1,n-1):
                d2udx2 = (Un[i+1,j] - 2. * Un[i,j] + Un[i-1,j]) * bydhsqr
                d2udy2 = (Un[i,j+1] - 2. * Un[i,j] + Un[i,j-1]) * bydhsqr
                d2vdx2 = (Vn[i+1,j] - 2. * Vn[i,j] + Vn[i-1,j]) * bydhsqr
                d2vdy2 = (Vn[i,j+1] - 2. * Vn[i,j] + Vn[i,j-1]) * bydhsqr
                uvsqr = Un[i,j]*(Vn[i,j]**2)
                
                U[i,j] = Un[i,j] + dt * (Du * (d2udx2 + d2udy2) - uvsqr + F * (1. - Un[i,j]))
                V[i,j] = Vn[i,j] + dt * (Dv * (d2vdx2 + d2vdy2) + uvsqr - (F + k) * Vn[i,j])
        
        
        #applying boundary conditions
        # edges
        
        U[0,:] = U[1,:]
        U[-1,:] = U[-2,:]
        U[:,0] = U[:,1]
        U[:,n-1] = U[:,n-2]
        
        V[0,:] = V[1,:]
        V[-1,:] = V[-2,:]
        V[:,0] = V[:,1]
        V[:,n-1] = V[:,n-2]
                    
    return (U,V)
```

```python
(Unew,Vnew) = ftcs(U, V, n, Du, Dv, F, k, nt, dt, dh)
```

```python
fig = plt.figure(figsize=(8,5))
plt.subplot(121)
plt.imshow(Unew, cmap = cm.RdBu)
plt.xticks([]), plt.yticks([]);
plt.subplot(122)
plt.imshow(Vnew, cmap = cm.RdBu)
plt.xticks([]), plt.yticks([]);
```
